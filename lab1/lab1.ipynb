Liczba przestaje być znormalizowana gdy pierwszy pierwszy bit zmienia się na 0 (przy potędze 2^-126). Ilość bitów cechy nie pozwala uzyskać niższej liczby.

````julia
#zadanie 4
#e^x = 1 + x + x^2/2! + x^3/3! + ... = ∑ x^i/i!; i∈{1,+∞} 
#dla x < 0 lepiej policzyć 1/(e^x), aby uniknąć problemu cancellation czyli odejmowania bliskich liczb

function pow_e_x(x, n)
    s1, s2, f, px = 1, 1, 1, abs(x)

    for i in 1:n
        temp = f*i
        s1 = s1 + x^i/temp
        s2 = s2 + px^i/temp
        f = temp
    end

    r1, r2 = 1/s1, 1/s2
    println("wartość w julii: ", exp(1)^x)
    println("niestabilne: ", r1, ", różnica: ", abs(r1 - exp(1)^x))
    println("stabilne: ", r2, ", różnica: ", abs(r2 - exp(1)^x))
end

pow_e_x(-10, 30)
````

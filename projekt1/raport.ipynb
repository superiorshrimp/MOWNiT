```julia
#libraries
using DataFrames
using CSV
using Plots
```

```julia
#projekt symulacji epidemii za pomocą CA zaimplementowałem w javie
#tutaj przedstawię przykładowe wyniki dla różnych parametrów początkowych
#opis parametrów:
#   immunity - losowy double z przedziału 0.0 do 1.0 opisujący siłę układu odpornościowego
#   transmission rate - łatwość z jaką wirus się rozprzestrzenia
#   decay - jak szybko odporność spada będąc zarażonym (odporność < 0.0 = śmierć)
#   immunity boost - boost odporności po przejściu choroby
#wywołania przy różnych parametrach wrzucone są na YT
#program javy zapisuje statystyki do .csv i poniżej rysowane są do tych plików wykresy 
```

```julia
#wywołanie 1:
#transmission rate = 0.1
#decay = 0.01
#immunity boost = 75
#https://youtu.be/dmxS4CniPaM

df = CSV.read("raport1.csv", delim = ",", DataFrame)

xs = [i for i in 1:nrow(df)]
scatter([xs xs xs], [df[!,1] df[!,2] df[!,3]],
label = ["healthy" "infected" "dead"], colors = ["GREEN", "YELLOW", "RED"],
title = "wywołanie 1.", xlab = "days", ylab = "people")
```

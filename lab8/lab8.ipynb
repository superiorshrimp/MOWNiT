```julia
using Roots  
using Plots
using ForwardDiff
using DataFrames
```

```julia
f1(x) = x^3
f2(x) = cos(x) - x
f3(x) = x^3 - 2*x - 5
f4(x) = x * exp(-x)
f5(x) = 2 * x - exp(-x)
f6(x) = sin(x) - x/2
print("")
```

```julia
df = DataFrame(alg = String[], f = String[], i = Int[])
```

```julia
plot(f1, -10:0.01:10)
```

```julia
#1
print("1: ")
x = find_zero(f1, (-1, 1), Bisection(), verbose = true)
println(iszero(f1(x)))

#2
print("2: ")
D(f1) = x -> ForwardDiff.derivative(f1, float(x))
x = find_zero((f1, D(f1)), eps(0.1), Roots.Newton(), verbose = true)
println(iszero(f1(x)))

#3
print("3: ")
x = find_zero(f1, 0, verbose = true)
println(iszero(f1(x)))

#
push!(df, ["bisection", "f1", 3])
push!(df, ["derivative", "f1", 2])
push!(df, ["der. approx.", "f1", 4])
```
